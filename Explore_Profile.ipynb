```python
"""

Explore profiling results

"""

import csv

FNAME = "profile.csv"
# COLS = ['   ncalls', '  tottime', '  percall', '  cumtime', ' filename:lineno(function)']
```

```python
row_to_metric = dict()
with open(FNAME) as csvfile:
    rows = list(csv.DictReader(csvfile))
    rows = sorted(rows, key=lambda row: -1 * float(row['  cumtime']))
    for i, row in enumerate(rows):
        print(f"{row['   ncalls']} \t {row['  cumtime']} \t {row[' filename:lineno(function)']}")
        if i == 100:
            break
    
```
