```python
import ast
import matplotlib.pyplot as plt
import os
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:80% !important; }</style>"))

# Change default plot size
plt.rcParams['figure.figsize'] = [22, 6]

logs = [
    "logs/cartpole_swingup/no_ss/noaugs/train.log",
    "logs/cartpole_swingup/no_ss/noise2net/train.log"
]
```

```python

# Plot test error vs epoch
def plot_col(metric):
    for log in logs:
        iters = []
        col_values = []
        with open(log) as f:
            while True:
                line = f.readline()
                if line == "":
                    break
                line = ast.literal_eval(line)    
                col_values.append(line[metric])
                iters.append(line['step'])
        plt.plot(iters, col_values, label=log)
```

```python
metric = "episode_reward"
plot_col(metric)
plt.legend(loc='lower right')
plt.title(metric)
plt.show()
```
